### Here is the function that returns the lattitude and longitude

```python
def geoGrab(address):
    from geopy.geocoders import Nominatim
    geolocator = Nominatim(user_agent = 'myApp')
    location = geolocator.geocode(address)
    if location is not None:
        loc = {}
        loc['latitude'] = location.latitude
        loc['longitude'] = location.longitude
        return loc
    else:
        return None
```

### Fill out this function placeFind()

```python
def placeFind(fileName):
    fw = open('places.txt', 'a')
    count = 0
    for line in open(fileName).readlines():
        count += 1
        # read rows and take their exact address 
        address = line.split("	")[1].strip()      
        # use geoGrab which returns a location stored in a dictionary format
        loc =  geoGrab(address)
        if loc != None:
            # write the longitude and latitude to places.txt
            lat = loc['latitude']
            long = loc['longitude']
            fw.write(f"{lat},{long}\n")
            
        else:
            # inform about error
            print(f"An error occurred in the line {count}")
        sleep(2) # the time pause needed as it takes some time to 
                 # connect through Google API and find location
    fw.close()

```

### Write your own function k-means:

```python
x = []
for line in open("places.txt").readlines():
    m = line.strip().split(',')
    x.append([float(m[0]), float(m[1])])
x = np.array(x) 
```

```python
def distance(c, x):
    return np.sqrt(np.square(x[0]-c[0])+ np.square(x[1]-c[1]))
```

```python
def clusterPlaces(numClust):
    # left this function empty as your code will be individual
    random_init_cent = x[np.random.choice(len(x), numClust)]
    cluster = [-1]*len(x)
    for i in range(2000):
# assign index of cluster
        for j in range(len(x)):
            dist_arr = []
            for k in range(numClust):
                dist_arr.append(distance(x[j], random_init_cent[k]))
            idx = np.argmin(dist_arr)
            cluster[j] = idx
# change centeroids by mean            
        random_init_cent = []
        for j in range(numClust):
            arr = []
            for k in range(len(x)):
                if cluster[k]==j:
                    arr.append(x[k])
            random_init_cent.append(np.mean(arr, axis=0))
#             print(random_init_cent)
    return cluster
    
```

At the end you need to obtain html file with points on the maps

```python
colors = {0:'blue', 1:'orange', 2:'green', 3:'pink'}
```

```python
cluster = clusterPlaces(4)
```

```python
my_map = folium.Map(location=[x[0][0], x[0][1]], zoom_start=12)
for i in range(len(x)):
     folium.Marker(location=[x[i][0], x[i][1]], icon=folium.Icon(color=colors[cluster[i]])).add_to(my_map)
my_map
```
