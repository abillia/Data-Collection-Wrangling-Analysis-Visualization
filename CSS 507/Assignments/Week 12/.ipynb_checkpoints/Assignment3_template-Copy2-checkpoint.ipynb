#### Hypothesis and cost fucntions

```python
x = np.c_[np.ones((x.shape[0], 1)), x]
x
```

```python
m = len(x)
alpha = 0.02
labels = np.array([10,  1,  2,  3,  4,  5,  6,  7,  8,  9])
t = np.zeros([x.shape[1], len(labels)])
print(labels)
print(t)
print(t.shape)
```

```python
def hypothesis(theta):
    z = np.dot(x, theta)
    return 1/(1+np.exp(-(z)))
h = hypothesis( t[:, 0])
```

```python
def cost_function(theta):
    y_pred = hypothesis(theta)
    return -(1/m) * np.sum(y*np.log(y_pred) + (1-y)*np.log(1-y_pred))
```

#### Gradient descent algorithm

```python
def gradient_descent(alpha, theta, y_tr):
    for j in range(0, len(theta)):
            theta[j] = theta[j] - alpha * (1/m)* np.sum((hypothesis(theta) - y_tr) * x[:,j])
            print(theta[j] )
    return theta
```

```python
y_tr = np.zeros([x.shape[0], len(labels)])
```

#### One vs all classification

```python
def One_vs_all(y, label):
    for i in range(len(label)):
        for j in range(len(y)):
            if y[j] == label[i]:
                y_tr[j, i] = 1
            else: 
                y_tr[j, i] = 0
```

```python
y_tr
```

```python
def predict(theta):
    y_hat = [0]*len(y)
    for i in range(len(labels)):
        y_t = y_tr[:, i]
        t = theta[:, i]
        y1 = hypothesis(t)
        for k in range(len(y)):
            if y_tr[k] == 1 and y1[k] >= 0.5:
                y_hat[k] = labels[i]
    return y_hat
```

#### Calculate accuracy

```python
y_pred = predict(t)
```

```python
accuracy=0
for i in range(0, len(y)):
    if y_pred[i] == y.flatten()[i]:
        accuracy += 1
print(accuracy/5000*100)
```
