__6. The dataset has elements of categorical data in the “Sport” column. Create a new dataframe with convertion of Sport column by one hot encoded column and left join this dataset to our main table.__

__Bonus: Find interesting information about Olympic games by analyzing a data with pandas queries__

__In this data frame, you can see how many medals each country has won during the entire Olympic Games, with the addition of the number of gold, silver, and bronze medals given in the dataset.__

```python
s = new_df.groupby(['Country', 'Medal']).agg({'Medal': pd.Series.count})
table = pd.DataFrame(s).unstack().fillna(0)
table2= new_df.groupby(['Country'], as_index=False).agg({'Medal': pd.Series.count}).rename(columns={'Medal': 'Total_Medal'})
```

```python
table.index
```

```python
# merging two tables the table and table2
data = table2.merge(table, left_on='Country', right_index=True)
data.sort_values('Total_Medal', ascending=False)
```
